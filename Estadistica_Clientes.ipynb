```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pedimos la cantidad de clientes en cada categoría
arba_web = int(input("¿Cuántos clientes tienes en Arba Web? "))
arba_rs = int(input("¿Cuántos clientes tienes en Arba RS? "))
convenio_multilateral = int(input("¿Cuántos clientes tienes en Convenio Multilateral? "))
agip_regimen_simplificado = int(input("¿Cuántos clientes tienes en Agip Regimen Simplificado? "))
responsables_inscriptos = int(input("¿Cuántos clientes tienes en Responsables Inscriptos? "))
doctores = int(input("¿Cuántos clientes tienes Doctores? "))

```

```python
# Creamos el DataFrame
df = pd.DataFrame({
    'Categoría': ['Arba Web', 'Arba RS', 'Convenio Multilateral', 'Agip Regimen Simplificado',
                  'Responsables Inscriptos', 'Doctores'],
    'Cantidad': [arba_web, arba_rs, convenio_multilateral, agip_regimen_simplificado,
                 responsables_inscriptos, doctores]
})

# Creamos el gráfico circular
plt.pie(df['Cantidad'], labels=df['Categoría'], autopct='%1.1f%%')
plt.title('Porcentajes de clientes en cada categoría')
plt.show()
```
